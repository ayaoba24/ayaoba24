```python
# Age Calculator - Calculate age from date of birth

from datetime import datetime, date

def calculate_age():
    try:
        # Get date of birth
        dob_input = input("Enter your date of birth (YYYY-MM-DD): ")
        dob = datetime.strptime(dob_input, "%Y-%m-%d").date()

        # Get today's date
        today = date.today()

        # Calculate age
        age_years = today.year - dob.year

        # Adjust if birthday hasn't occurred this year
        if (today.month, today.day) < (dob.month, dob.day):
            age_years -= 1

        # Calculate total days
        total_days = (today - dob).days

        # Calculate months
        months = (today.year - dob.year) * 12 + (today.month - dob.month)
        if today.day < dob.day:
            months -= 1

        print(f"\nAge Information:")
        print(f"You are {age_years} years old")
        print(f"Total months lived: {months}")
        print(f"Total days lived: {total_days}")

    except ValueError:
        print("Error: Please enter date in YYYY-MM-DD format!")
    except Exception as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    calculate_age()
```
