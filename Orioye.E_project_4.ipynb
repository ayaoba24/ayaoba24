```python
# Step 1: Import necessary modules
import time  # Gives access to time-related tools
import winsound  # To make beep sound at the end
from tqdm import tqdm  # To show progress bar
import msvcrt  # For detecting key presses (works on Windows only)

# Step 2: User inputs the countdown duration
seconds = int(input("Enter time in seconds: "))

# Step 3: Set initial pause state
paused = False
i = seconds  # Countdown counter

# Step 4: Start the countdown loop with tqdm for progress bar
for _ in tqdm(range(seconds, 0, -1)):  # tqdm shows progress graphically
    if msvcrt.kbhit():  # Check if any key is pressed
        key = msvcrt.getch().decode().lower()
        if key == 'p':
            paused = True
            print("Paused. Press 'r' to resume.")
        elif key == 'r':
            paused = False
            print("Resumed.")
        elif key == 'q':
            print("Timer stopped early.")
            break

    if not paused:
        mins = i // 60  # Convert seconds to minutes
        secs = i % 60   # Get remaining seconds
        print(f"Time left: {mins}m {secs}s")  # Better readability
        time.sleep(1)
        i -= 1
    else:
        time.sleep(1)  # Still wait while paused

# Step 5: Final alert after countdown ends
if i == 0:
    print("Time up!")  # Notify user
    winsound.Beep(1000, 1000)  # (frequency, duration in milliseconds)




```
