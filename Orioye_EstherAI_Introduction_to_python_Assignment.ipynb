### Part 3: Basic Python Operations
- Given two numbers a = 15 and b = 4, perform the following operations and print the results:

  - Addition
  - Floor division
  - Exponentiation

- Write a Python program that swaps the values of two variables without using a third variable.

```python

a = 15
b = 14

# Addition: Adds a and b
print("Sum:", a + b)

# Floor division: Divides a by b and rounds down to the nearest whole number
print("Floor Division:", a // b)

# Exponentiation: Raises a to the power of b
print("Exponentiation:", a ** b)

# This technique is a concise way to swap variable values in Python
a, b = b, a

# Display the values of the variables after swapping
print("After Swapping: a =", a, "b =", b)
```

- The following code has an error. What error is it? try to fix it:

```python
my_tuple = (1, 2, 3)
my_tuple[0] = 10 # It has a TypeError: 'tuple' object does not support item assignment
```

```python
my_tuple = (1, 2, 3)

# Convert the original tuple to a list to allow modification
my_list = list(my_tuple)

# Modify the first element of the list
my_list[0] = 10
print(my_list) # Display the modified list

# Convert the modified list back into a tuple
my_tuple = tuple(my_list)
print(my_tuple) # Display the new tuple with the changed element
```
