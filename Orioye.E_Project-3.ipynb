```python
#Temperature Converter
"""
Key Concepts of Temperature Converter in Python

Mathematical Conversion Formulas:
- Celsius to Fahrenheit: (C × 9/5) + 32
- Fahrenheit to Celsius: (F - 32) × 5/9
- Celsius to Kelvin: C + 273.15
- Kelvin to Celsius: K - 273.15
- Fahrenheit to Kelvin: (F - 32) × 5/9 + 273.15
- Kelvin to Fahrenheit: (K - 273.15) × 9/5 + 32

"""

# Functions in Python:
# Defining functions for conversions
def c_to_f(c): return (c * 9/5) + 32
def f_to_c(f): return (f - 32) * 5/9
def c_to_k(c): return c + 273.15
def k_to_c(k): return k - 273.15
def f_to_k(f): return (f - 32) * 5/9 + 273.15
def k_to_f(k): return (k - 273.15) * 9/5 + 32

conversions = {
    '1': ("Celsius to Fahrenheit", c_to_f, "°C", -273.15, "({} × 9/5) + 32 = {:.2f}°F"),
    '2': ("Fahrenheit to Celsius", f_to_c, "°F", -459.67, "({} - 32) × 5/9 = {:.2f}°C"),
    '3': ("Celsius to Kelvin", c_to_k, "°C", -273.15, "{} + 273.15 = {:.2f} K"),
    '4': ("Kelvin to Celsius", k_to_c, "K", 0, "{} - 273.15 = {:.2f}°C"),
    '5': ("Fahrenheit to Kelvin", f_to_k, "°F", -459.67, "({} - 32) × 5/9 + 273.15 = {:.2f} K"),
    '6': ("Kelvin to Fahrenheit", k_to_f, "K", 0, "({} - 273.15) × 9/5 + 32 = {:.2f}°F")
}

while True:
    print("\nSelect conversion:")
    for k, (label, *_rest) in conversions.items():
        print(f"{k}. {label}")
    print("0. Exit")

    choice = input("Enter choice (1-20 or 0): ")
    if choice == '0': break
    if choice not in conversions:
        print("Invalid choice.")
        continue

    label, func, unit, min_temp, formula = conversions[choice]
    print(f"\n{label} selected. Type 'back' to reselect or '0' to exit.")

    while True:
        val = input(f"Enter temperature in {unit}: ")
        if val.lower() == 'back': break
        if val == '0': exit()
        try:
            temp = float(val)
            if temp < min_temp:
                print("Below absolute zero. Try again.")
                continue
            result = func(temp)
            print(formula.format(temp, result))
        except ValueError:
            print("Invalid input. Enter a number or command.")

```
