```python
from collections import defaultdict

# Helper function to find all empty positions
def find_empty_positions(grid):
    return [(r, c) for r in range(9) for c in range(9) if grid[r][c] == 0]

# Helper function to compute possible values for each cell
def compute_possibilities(grid):
    possibilities = defaultdict(set)
    for row in range(9):
        for col in range(9):
            if grid[row][col] == 0:
                used = set(grid[row])  # Row values
                used.update(grid[i][col] for i in range(9))  # Column values
                # Sub-grid values
                start_row, start_col = 3 * (row // 3), 3 * (col // 3)
                used.update(grid[i][j] for i in range(start_row, start_row + 3)
                                        for j in range(start_col, start_col + 3))
                possibilities[(row, col)] = set(range(1, 10)) - used
    return possibilities

# Constraint Propagation + Backtracking Solver
def solve_with_constraints(grid):
    empty_positions = find_empty_positions(grid)

    def backtrack():
        if not empty_positions:
            return True

        # Recompute possibilities to keep them updated
        possibilities = compute_possibilities(grid)

        # Pick the position with the fewest possibilities (MRV heuristic)
        row, col = min(empty_positions, key=lambda pos: len(possibilities[pos]))
        possible_nums = possibilities[(row, col)]

        for num in possible_nums:
            grid[row][col] = num
            empty_positions.remove((row, col))

            if backtrack():
                return True

            # Undo the move
            grid[row][col] = 0
            empty_positions.append((row, col))

        return False

    return backtrack()

# Function to print the Sudoku grid
def print_grid(grid):
    for row in range(9):
        for col in range(9):
            print(grid[row][col], end=" ")
            if (col + 1) % 3 == 0 and col < 8:
                print("|", end=" ")
        print()
        if (row + 1) % 3 == 0 and row < 8:
            print("-" * 21)

# The Sudoku puzzle
grid = [
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9]
]

# Solve and print
if solve_with_constraints(grid):
    print_grid(grid)
else:
    print("No solution exists")
```
