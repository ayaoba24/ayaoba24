```python

import re
import json
import streamlit as st

# ------------------- Madlibs Data --------------------
madlibs_data = {
    # Theme: "Simple Variety"
    "Simple Variety": {
        # Story title as key
        "CMD PMT Chicken: control thou pecker": {
            "glimpse": '''
            This lil man done just spoke... Chicken??!!!
            'cluuuck' 'cluck' 'cluck'
            into the coop the fowls go.
            Just as i turn to _____(k_verb)_____ the last one who's just _____(l_verb)_____, looking around,
            i look a little to my _____(m_direction)_____ and
            BOOM
            ''',
            "story": '''
            While looking out my window, trying to savor a little bit of serendipity,
            the neighbors son, Sola, taunted Mr Dany's fowls with the really hot noodles he was munching on.
            How that young man is able to down such hot foods i'd never know.
            What caught my eye next was this shiny dark crow circling overhead,
            once at each pole of its wide, invisible orbit, it cried out.
            It did this twice,
            and each time, Mayana, the Cuckling hen, cried out in response.
            To end this poutric interaction, all the chicken calmly, yet quickly rushed into their coop.
            At the moment, i thought nothing of it, until...
            I go out to open the gate for my bro;
            Now these chicken brains won't clear out the way and chasing them?
            Come on, my steeze.
            I was about sacrificing my precious grondnuts to lure them out of the way when
            'cluuuck' 'cluck' 'cluck'
            into the coop the fowls go.
            Just as i turn to kick the last one who's just wandering looking around,
            i look a little to my left and
            BOOM
            Sola grinning ear-to-ear
            That's when i recognised the clucking just now and it wasn't Mayana
            It was...
            You guessed it
            Sola
            This lil man done just spoke... Chicken??!!!
            ''',
            "madlibs": '''
            While _____(a_verb)_____ out my window, trying to _____(b_verb)_____ a little bit of _____(c_noun)_____,
            the neighbors son, Sola, taunted Mr Dany's fowls with the really hot _____(d_food)_____ he was munching on.
            How that young man is able to down such hot foods i'd never know.
            What _____(e_verb)_____ my eye next was this shiny dark crow circling overhead,
            once at each pole of its wide, invisible orbit, it cried out.
            It did this twice,
            and each time, Mayana, the Cuckling hen, cried out in response.
            To end this poutric interaction, all the chicken calmly, yet quickly rushed into their _____(f_place)_____.
            At the moment, i thought nothing of it, until...
            I go out to _____(g_verb)_____ the gate for my bro;
            Now these chicken brains won't clear out the way and chasing them?
            Come on, my _____(h_noun)_____.
            I was about sacrificing my precious _____(i_snack)_____ to _____(j_verb)_____ them out of the way when
            'cluuuck' 'cluck' 'cluck'
            into the coop the fowls go.
            Just as i turn to _____(k_verb)_____ the last one who's just _____(l_verb)_____, looking around,
            i look a little to my _____(m_direction)_____ and
            BOOM
            Sola grinning ear-to-ear
            That's when i recognised the clucking just now and it wasn't _____(n_noun)_____
            It was...,
            You guessed it,
            _____(o_noun)_____
            This lil man done just spoke... Chicken??!!!
            '''
        }
    },

    # Theme: "Themed (Fantasy, Adventure)"
    "Themed (Fantasy, Adventure)": {
        "Echoes of the Fallen Star: Necromancer's comet": {
            "glimpse": '''
            'clank' 'clank' 'clank'
            'where am i?' 'where is this?'
            Yellings and screams coming from the rearguard.
            'oh what?, i saw that guy go down'
            'yea, i already ended that one too'
            'i put one through his helmet, i'm sure of that
            'retreat! retreat! retreat!', ordered the general
            ''',
            "story": '''
            In an Icy Wasteland, right in the middle of a raging blizzard,
            a determined swordsmith transverses the unforgiving terrain in search the endpoint of the bright dot.
            He watching blazzing across the sky couple nights ago while engraving his latest piece.
            Here was his moment to seetup a leagcy as the greatest swordsmith in the continent,
            or as the fool who willingly ran to his tomb and sealed himself in.
            However, if a succeeds, and is able to get even just a piece of the 'dropping rock',
            this would be a great game-changer for the kingdom,
            swords sharper, more durable, lighter than any iron.
            Through the unforgiving cold, he races across the icy trail leading to the crash site.
            'clank' 'clank' 'clank', he chips away.
            'what is all that?' 'what is all that noise?' 'what is happening?'
            he pauses,
            'must be the howling winds', he sighed
            'clank' 'clank' 'clank'
            'where am i?' 'where is this?'
            ''Arghh', no, no, no, i'm out of here', he bellowed as he gathered what he could carry on hand and ran out,
            dropping little pieces on the bare trail back out.
            **********************************************************************************************
            'Blades incomparable to any steel out there, with a poisoning effect despite severity of injury',
            reported the kingdom's army general to the council.
            'Prepare the troops', ordered the council head.
            **********************************************************************************************
            On the battle field, the troops are able to push enemy lines, killings by the wagonload.
            'Victory! Victory! Victory!', chanted the troops.
            'what is all that?' 'where is it coming from?' 'what is happening?'
            Just at the precipice of victory
            Yellings and screams coming from the rearguard
            'oh what?, i saw that guy go down'
            'yea, i already ended that one too'
            'i put one through his helmet', i'm sure of that
            'Retreat! Retreat! Retreat!', ordered the general
            **********************************************************************************************
            'my lords, we had to retreat, the battle took a strange turn, we had no choice',
            reported the genral.
            'Understandable, but why do you have the troops surrounding us, with their weapons out', council_chief1
            'I command you put away your weapons', council_chief2
            'Control your troops, general', ordered the council head
            The general retorted, 'I'm sorry but we do answer to you'.
            'We answer to a higher command', echoed the troops as they pounced on the council.
            **********************************************************************************************
            END.
            ''',
            "madlibs": '''
            In an Icy _____(a_location)_____, right in the middle of a raging _____(b_weather)_____,
            a _____(c_adjective)_____ swordsmith transverses the unforgiving terrain in search the endpoint of the bright dot.
            He watched blazzing across the sky couple nights ago while engraving his latest _____(d_noun)_____.
            Here was his moment to setup a leagcy as the greatest swordsmith in the _____(e_location)_____,
            or as the fool who willingly ran to his _____(f_place)_____ and _____(g_verb)_____ himself in.
            However, if he succeeds, and is able to get even just a piece of the 'dropping rock',
            this would be a great _____(h_noun)_____ for the kingdom,
            swords sharper, more _____(i_adjective)_____, lighter than any iron.
            Through the unforgiving _____(j_adjective)_____, he _____(k_verb)_____ across the icy trail leading to the _____(l_noun)_____ .
            'clank' 'clank' 'clank', he _____(m_verb)_____ away.
            'what is all that?' 'what is all that noise?' 'what is happening?'
            he pauses,
            'must be the howling winds', he sighed
            'clank' 'clank' 'clank'
            'where am i?' 'where is this?'
            'Arghh', no, no, no, i'm out of here, he bellowed as he gathered what he could carry on hand and ran out,
            dropping little pieces on the bare trail back out.
            **********************************************************************************************
            'Blades _____(n_adjective)_____ to any steel out there, with a _____(o_adjective)_____ effect despite _____(p_adjective)_____ of injury',
            reported the kingdom's army general to the council.
            'Prepare the troops', ordered the council head.
            **********************************************************************************************
            On the battle field, the troops are able to push enemy lines, _____(q_verb)_____ by the _____(r_noun)_____.
            'Victory! Victory! Victory!', chanted the troops.
            'what is all that?', 'where is it coming from?', 'what is happening?'
            Just at the _____(s_noun)_____ of victory.
            Yellings and screams coming from the rearguard.
            'oh what?, i saw that guy go down'
            'yea, i already ended that one too'
            'i put one through his helmet', i'm sure of that
            'retreat! retreat! retreat!', ordered the general
            **********************************************************************************************
            'my lords, we had to _____(t_verb)_____, the battle took a _____(u_adjective)_____ turn, we had no _____(v_noun)_____',
            reported the genral.
            'Understandable, but why do you have the troops _____(w_verb)_____ us, with their _____(x_noun)_____ out', council_chief1
            'I command you put away your _____(y_noun)_____', council_chief2
            'Control your _____(z_noun)_____, general', ordered the council head
            The general _____(a_verb)_____, 'I'm sorry but we do not _____(b_verb)_____ to you'.
            'We answer to a _____(c_verb)_____ command', _____(d_verb)_____ the troops as they _____(e_verb)_____ on the council.
            **********************************************************************************************
            END.
            '''
        }
    }
}

# ------------------- Helper Functions --------------------

def get_placeholders(template):
    placeholders = re.findall(r'_\((\w+)\)_', template)
    unique_placeholders = sorted(set(placeholders))
    return unique_placeholders


def show_progress(template, user_answers):
    def replacement(match):
        tag = match.group(1)  # get placeholder tag inside _( )_
        if tag in user_answers and user_answers[tag].strip() != '':
            return user_answers[tag]  # replace with user answer if present and non-empty
        else:
            return f"[{tag}]"  # else show placeholder tag for visibility
    completed_text = re.sub(r'_\((\w+)\)_', replacement, template)
    return completed_text

# -------- Function to fill a specific placeholder with user input -----------------
def fill_placeholder(user_answers, placeholders, tag):
    if tag not in placeholders:
        print(f"Placeholder '{tag}' not found.")
        return
    answer = input(f"Enter a word or phrase for placeholder '{tag}': ").strip()
    user_answers[tag] = answer
    print(f"Filled '{tag}' with: {answer}")

# -------- Function to fill all placeholders one by one ------------------------
def fill_all_placeholders(user_answers, placeholders):
    for tag in placeholders:
        if tag in user_answers and user_answers[tag].strip() != '':
            continue
        fill_placeholder(user_answers, placeholders, tag)

# -------- Function to save progress to a JSON file -------------------------------
def save_progress(story_title, user_answers):
    filename = story_title.replace(' ', '_').replace(':', '') + '_progress.json'
    try:
        with open(filename, 'w', encoding='utf-8') as f:
            json.dump(user_answers, f, ensure_ascii=False, indent=2)
        print(f"Progress saved to '{filename}'.")
    except Exception as e:
        print(f"Error saving progress: {e}")

# -------- Function to load progress from a JSON file -----------------------------
def load_progress(story_title):
    filename = story_title.replace(' ', '_').replace(':', '') + '_progress.json'
    try:
        with open(filename, 'r', encoding='utf-8') as f:
            answers = json.load(f)
        print(f"Progress loaded from '{filename}'.")
        return answers
    except FileNotFoundError:
        print(f"No saved progress found for '{story_title}'. Starting fresh.")
        return {}
    except Exception as e:
        print(f"Error loading progress: {e}")
        return {}
# ------------------- Streamlit App --------------------

st.title("📝 Mad Libs Game")

# Theme selection
themes = list(madlibs_data.keys())
selected_theme = st.selectbox("Choose a theme:", themes)

# Story selection
stories = list(madlibs_data[selected_theme].keys())
selected_story = st.selectbox("Choose a story:", stories)

story_data = madlibs_data[selected_theme][selected_story]
glimpse = story_data["glimpse"]
template = story_data["madlibs"]
placeholders = get_placeholders(template)

# Load progress
if 'answers' not in st.session_state:
    st.session_state.answers = load_progress(selected_story)

st.markdown("### ✨ Glimpse")
st.info(glimpse.strip())

st.markdown("### 🧩 Fill in the blanks")
with st.form("fill_form"):
    for tag in placeholders:
        default = st.session_state.answers.get(tag, "")
        st.session_state.answers[tag] = st.text_input(f"{tag}", value=default)
    submitted = st.form_submit_button("Update")

# Buttons for save, load, complete view
col1, col2, col3, col4, col5, col6, col7 = st.columns(7)

if col1.button("💾 Save Progress"):
    save_progress(selected_story, st.session_state.answers)
    st.success("Progress saved!")

if col2.button("📂 Load Progress"):
    st.session_state.answers = load_progress(selected_story)
    st.success("Progress loaded!")

if col3.button("📜 Show Completed Story"):
    completed_story = show_progress(template, st.session_state.answers)
    st.markdown("### ✅ Completed Story")
    st.text_area("Your Story:", completed_story, height=600)

if col4.button("📖 Show original story"):
    st.subheader("Original Story")
    st.text(story_data["story"])

if col5.button("📥 Download progress"):
    filename = f"{selected_story}_progress.json"
    json_data = json.dumps(st.session_state.answers, indent=2)
    st.download_button("Download Progress", json_data, file_name=filename)

if col6.button("📄 Download original story"):
    filename = f"{selected_story}_original.json"
    json_data = json.dumps(story_data, indent=2)
    st.download_button("Download Original Story", json_data, file_name=filename)

if col7.button("🚪 Quit"):
    st.subheader("Thanks for playing Mad Libs! 👋")
    st.stop()
```
